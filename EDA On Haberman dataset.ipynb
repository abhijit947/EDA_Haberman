# Univariate Analysis


# 1-D Scatter Plot


```python
import numpy as np

haberman_1=haberman.loc[haberman['status']==1]
haberman_2=haberman.loc[haberman['status']==2]


plt.plot(haberman_1['age'],np.zeros_like(haberman_1['age']),'o',label='status 1')
plt.plot(haberman_2['age'],np.zeros_like(haberman_2['age']),'o',label='status 2')
plt.grid()
plt.legend()
plt.title('1-D Scatter Plot of age of patients ')
plt.show()
```

```python
plt.plot(haberman_1['year'],np.zeros_like(haberman_1['year']),'o',label='status 1')
plt.plot(haberman_2['year'],np.zeros_like(haberman_2['year']),'o',label='status 2')
plt.grid()
plt.legend()
plt.title('1-D Scatter Plot of year of operation of patients ')
plt.show()
```

```python
plt.plot(haberman_1['nodes'],np.zeros_like(haberman_1['nodes']),'o',label='status 1')
plt.plot(haberman_2['nodes'],np.zeros_like(haberman_2['nodes']),'o',label='status 2')
plt.grid()
plt.legend()
plt.title('1-D Scatter Plot of number of axillary lymph nodes ')
plt.show()
```

# Observation
After plotting 1-D scatter plot of age,year and nodes,it is observed that nothing can be concluded as status 1 and status 2 are overlapping.

# Histogram

```python
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(sns.distplot,'year') \
    .add_legend();
plt.title("Histogram of year of operation of patient")
plt.grid()
plt.show()
```

```python
sns.FacetGrid(haberman,hue='status',size=5) \
    .map(sns.distplot,'nodes')\
    .add_legend()
plt.title("Histogram of number of axillary lymph nodes")
plt.grid()
plt.show()
```

# Observation
After plotting Histogram of age,year and nodes, we found that status 1 and status 2 are overlapping and hence nothing can be concluded from these analysis.

# PDF and CDF

```python
# For status 1
count,bin_edges=np.histogram(haberman_1['age'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 1:pdf')
plt.plot(bin_edges[1:],cdf,label='status 1:cdf')
bin_edges

# For status 2
count,bin_edges=np.histogram(haberman_2['age'],bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 2:pdf')
plt.plot(bin_edges[1:],cdf,label='status 2:cdf')
plt.legend()
plt.title('CDF and PDF of age of patients')
plt.grid()
plt.show()
```

```python
#for status 1
count,bin_edges=np.histogram(haberman_1['year'],bins=9)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 1:pdf')
plt.plot(bin_edges[1:],cdf,label='status 1:cdf')

#for status 2
count,bin_edges=np.histogram(haberman_2['year'],bins=9)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 2:pdf')
plt.plot(bin_edges[1:],cdf,label='status 2:cdf')
plt.legend()
plt.title('CDF and PDF of year of operation of patients')
plt.grid()
plt.show()
```

```python
# For status 1
count,bin_edges=np.histogram(haberman_1['nodes'],bins=9)
print(count)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 1:pdf')
plt.plot(bin_edges[1:],cdf,label='status 1:cdf')

# For status 2
count,bin_edges=np.histogram(haberman_2['nodes'],bins=9)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='status 2:pdf')
plt.plot(bin_edges[1:],cdf,label='status 2:cdf')
plt.title('CDF and PDF of number of axillary lymph nodes')
plt.legend()
plt.grid()
plt.show()
```

# Observation
After plotting PDF and CDF of age,year and nodes we found that status 1 and status 2 are overlapping and hence nothing can be concluded from these analysis.


# Boxplot

```python
sns.boxplot(x='status',y='age',data=haberman)
plt.grid()
plt.title('Box plot of age and status of patients')
plt.show()
```

```python
sns.boxplot(x='status',y='year',data=haberman)
plt.grid()
plt.title('Box plot of year of operation and status of patients')
plt.show()
```

```python
sns.boxplot(x='status',y='nodes',data=haberman)
plt.grid()
plt.title('Box plot of number of axillary lymph nodes and status of patients')
plt.show()
```

# Violin Plots

```python
sns.violinplot(x='status',y='age',data=haberman,size=10)
plt.title("violin plot of age of patients and status of patients")
plt.grid()
plt.show()
```

```python
sns.violinplot(x='status',y='year',data=haberman,size=10)
plt.title("violin plot of year of operation of patients and status of patients")
plt.grid()
plt.show()
```

```python
sns.violinplot(x='status',y='nodes',data=haberman,size=10)
plt.title("violin plot of number of axillary lymph nodes and status of patients")

plt.grid()
plt.show()
```

# Observation
From the box plot and violin plot of age of the patient,it is observed that most of the patients whose age was between 30 and 35 years, survived for more than five years and patients whose age was above 75 years, survived for less than five years.

From the box plot and violin plot of year of operation of patients,it is observed that most of the patients whose operation was done before year 60, survived for less than five years and  most of the patients whose operation was done after year 65, survived for more than five years.

From the box plot and violin plot of number of axillary lymph nodes,it is obseved that only 25% of patients who had axillary lymph nodes more than 3, survived for more than five years.The higher the number of axillary lymph nodes,more is the chances of patients surviving less than five years.
